```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.tri as mtri
import scipy as sp
from scipy import optimize
from numpy import linalg
```

```python
def runge(f, n, x_1, x_2):
    I = 0; 
    for  i in range(n):
        I +=_runge(f,i*((x_2)-(x_1))/n, (i+1)*((x_2)-(x_1))/n)
    return I
def _runge(f,x_1,x_2):
    return (x_2 - x_1)/2. *(f((x_1 + x_2)/2. - (x_2 - x_1)/(2*np.sqrt(3.))) + f((x_1 + x_2)/2. + (x_2 - x_1)/(2.*np.sqrt(3.)))) 
        
```

```python
def f(x):
    return np.log(1/(np.sin(x)))
err = 1
eps = 10e-7
n =1
I_1 = 1
I_2 = 1
error = []
h_list = []
while (err>eps):     
    I_1 = runge(f,n, 0, np.pi/2)
    if (n == 1):
        n *=2
        continue
    err = np.abs(I_2 - I_1)
    I_2 = I_1 + 0.0    
    error.append(err)
    h = 1./n
    h_list.append(h)
    n *=2
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('h')
ax.set_ylabel('error')
plt.title('Without regularisation')
plt.loglog(h_list, error)
plt.show()

    
    
    
```
