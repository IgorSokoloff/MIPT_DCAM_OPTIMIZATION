# Задача №1 

Выпуклы ли следующие функции:

* $$f(x) = e^{x} - 1, x \in \Re$$
* $$f(x_1, x_2) = x_1 \cdot x_2, x \in \Re^2_{++}$$
* $$f(x_1, x_2) = \frac{1}{x_1 \cdot x_2}, x \in \Re^2_{++}$$





Решение: 

1) Вообще, для первой функции можно использовать тяжелую артиллерию и сразу применить к ней дифференциальный критерий 2-го порядка. Вторая производная от данной функции будет равна просто экспоненте $\Rightarrow$ она будет положительна для все значений $x$ и условия критерия выполнены - функция выпуклая. 

Но можно воспользоваться и критерием первого порядка: 

$$f(y) \ge f(x) + \nabla{f^T(x)}(y-x)$$

Начнем с того, что $\forall x \in \Re$ верно
$$e^x - 1 \ge x$$

Воспользуемся этим фактом для x = z - y
$$e^{(z-y)} - 1 \ge (z-y)$$
Домножим обе части на $e^y$ и вычтем $1$:
$$e^{z} - e^{x} - 1 \ge e^x(z-y) - 1$$
$$e^{z} - 1 \ge e^y - 1 + e^y (z - y)$$

Что и будет выражением, которое используется в критерии $\Rightarrow$ функция выпуклая и по второму критерию.


2) Воспользуемся дифференциальным критерием второго порядка: 

Гессиан для данной фунции запишется следующим образом:

$$Hess_{f}(x) = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$$
Данная матрица не является положительно\отрицательно определенной  $\Rightarrow $ - функция не является ни выпуклой вверх, ни вниз

3) Снова воспользуемся вторым диффенренциальным критерием для данной функции: 

$$Hess_{f}(x) = \begin{bmatrix} \frac{2}{x_1^3 x_2} & \frac{1}{x_1^2 x_2^2} \\ \frac{1}{x_1^2 x_2^2} & \frac{2}{x_2^3 x_1} \end{bmatrix}$$

Данная матрица будет положительно определенной $\forall x \in \Re^2_{++} \Rightarrow$ функция является выпуклой 

# Задача №2 

Докажите, что множество $S = \{x \in \Re^n \mid \prod_{i=1}^{n} x_i \ge 1 \} $ выпуклое

Решение: 

Представим данное множество в другом виде, а именно, прологорифмируем обе части неравенства.

$$S = {x \in \Re^n \mid \sum \limits_{i=1}^{n} \ln{x_i} \ge 0}$$

Зададимся некоторыми $x_1, x_2 \in S$ и некоторым  $\lambda \in [0, 1]$
В силу вогнутости функции  $\ln{x}$ можно записать неравенство вида:

$$\sum \limits_{i=1}^{n}\ln{(\lambda x_{1i} + (1-\lambda) x_{2i})} \ge \lambda \sum \limits_{i=1}^{n} \ln{x_{1i}} + (1-\lambda) \sum \limits_{i=1}^{n} \ln{x_{2i}} \ge 0 $$

Отсюда получаем, что множество выпуклое просто по определению.

# Задача №3 

Докажите, что функция $f(X) = \mathbf{tr}(X^{-1}), X \in S^{n}_{++}$ выпукла, а  $g(X) = (\det{X})^{\frac{1}{n}}, X \in S^{n}_{++}$ вогнута.

Решение: 

$\textbf{1)}$ Для первой функции воспользуемся тем результатом, что для любой симметричной положительно определеной матрицы существует базиси из собственных векторов, в котором она имеет диагональный вид

Тогда зададимся некоторыми матрицами 
$$X_1, X_2 \in S^{n}_{++}, \alpha \in [0, 1]$$
$$X^* = \alpha X_1 + (1-\alpha)X_2$$

Пусть для матриц $X^*, X_1, X_2$ матрицами перехода к базису из собственных векторов являются матрицы $S^*, S_1, S_2$; $D^*, D_1, D_2$ - матрицы, имеющие на диагонали собственные числа матриц $X^*, X_1, X_2$, тогда можно записать следующую штуку

$$\mathbf{tr}((X^*)^{-1}) = \mathbf{tr}((S^* \cdot D^* \cdot (S^*)^{-1})^{-1}) = \mathbf{tr}((\lambda(\alpha X_1 + (1-\lambda X_2)))^{-1})= \mathbf{tr}((\alpha \lambda(X_1) + (1-\alpha)\lambda(X_2))^{-1}) = \sum \limits_{i=1}^{n} \frac{1}{\alpha \lambda_{1i} + (1-\alpha) \lambda_{2i}}$$

где за $\lambda(X)$ - обозначаем взятие собственных чисел матрицы, а $\lambda_{1i}, \lambda_{2i}$ - соответствующие собственные числа матриц $X_1, X_2$

С другой стороны рассмотрим 
$$\alpha \cdot \mathbf{tr}((X_1)^{-1}) + (1 - \alpha) \cdot  \mathbf{tr}((X_2)^{-1}) \Rightarrow$$
Аналогично рассуждениям выше находим базисы из собсвенных векторов, к которых матрицы имею диагональный вид. Отсюда получаем:

$$\Rightarrow \alpha \sum \limits_{i=1}^{n} \frac{1}{\lambda_{1i}} + (1 - \alpha) \sum \limits_{i=1}^{n} \frac{1}{\lambda_{2i}}$$

Воспользуемся очевидной выпуклостью функции $\frac{1}{x}$, для нее запишем: 

$$ \sum \limits_{i=1}^{n} \frac{1}{\alpha \lambda_{1i} + (1-\alpha) \lambda_{2i}} \le \alpha \sum \limits_{i=1}^{n} \frac{1}{\lambda_{1i}} + (1 - \alpha) \sum \limits_{i=1}^{n} \frac{1}{\lambda_{2i}}$$

Но это неравенство в точности соответсвует неравенству вида: 

$$\mathbf{tr}((X^*)^{-1}) = \mathbf{tr}((\alpha X_1 + (1 - \alpha) X_2)^{-1}) \le \alpha \cdot \mathbf{tr}((X_1)^{-1}) + (1 - \alpha) \cdot  \mathbf{tr}((X_2)^{-1})$$

Что просто по определению говорит о том, что функиция выпуклая

$\textbf{2)}$ Для решения данной задачи воспользуемся некоторыми вспомогательными фактами. 

Для начала, вспомним, что функция вида 

$$f(X) = - \log{\det{X}} \mbox{ is convex on } \textbf{dom}(f(x))  = S^n_{++}$$

Теперь прологарифмируем данную функцию $g(x)$. Тем самым мы как раз и получим $\frac{1}{n}\log{\det{X}}$. Про нее мы уже знаем, что она вогнутая. Было бы круто сейчас просто пропотенцировать данную функцию и сказать, что наша вогнутость сохранилась.

И в этот момент к нам на помощь приходит теорема из данного [источника](https://ljk.imag.fr/membres/Anatoli.Iouditski/cours/convex/chapitre_3.pdf):

Под названием "Convex  Monotone  superposition", которая утверждает то, что если у нас есть некоторая вектор-функция $$f(x)\colon \mathbb{R^n} \to \mathbb{R^k}$$ 

которая имеет все выпуклые/вогнутые компоненты и есть некоторая функция 

$$F(x) \colon \mathbb{R^k} \to \mathbb{R}$$

которая является выпуклой/вогнутой и монотонно возрастающей/убывающей, то 

$$F(f(x))\colon \mathbb{R^n} \to \mathbb{R}$$

тоже является выпуклой/вогнутой.

Теперь внимательно посмотрим на наши функции и скажем, что 
$$F(x) = - e^{x}$$

$$f(x) = \frac{1}{n} \log{\det{X}}$$

$$F(f(x)) = -\det{X}^{\frac{1}{n}}$$

Так же будет вогнутая, что и хотели показать  $\Rightarrow (\det{X})^{\frac{1}{n}}$ - выпуклая функция 

P.S: Доказательство теоремы действительно довольно простое, нужно только немного подумать





# Задача №4

Расстоянием Кульбака-Лейблера между $p, q \in \mathbb{R}^{n}_{++}$ называется

$$D(p, q) = \sum \limits_{i=1}^{n} (p_i log(\frac{p_i}{q_i}) - p_i + q_i)$$

Докажите, что $D(p,q) \ge 0 \forall p, q \in \mathbb{R}^{n}_{++}$ и $D(p, q) = 0 \leftrightarrow p=q$

Решение: 

Для доказательства воспольуемся неравенством, которое называется ["Log sum inequality"](https://en.wikipedia.org/wiki/Log_sum_inequality):

$$\sum \limits_{i=1}^{n} a_i \log{\frac{a_i}{b_i}} \ge a \log{\frac{a}{b}}$$
$$\sum \limits_{i=1}^{n} a_i = a$$
$$\sum \limits_{i=1}^{n} b_i = b$$

Причем равенство в данном неравенстве выпоняется только в том случае, если $a_i = b_i, i = \overline{1,n}$

Так же нам надо вспомнить, что для логарифма верно, что

$$\ln{x} \le x - 1$$

А равенство будет верно только при $x=1$

Еще раз запишем определение расстояния Кульбака-Лейблера и воспользуемся для него неравенствами выше: 

$$D(p, q) = \sum \limits_{i=1}^{n} (p_i log(\frac{p_i}{q_i}) - p_i + q_i) \ge p log(\frac{p}{q}) + \sum \limits_{i=1}^{n} (- p_i + q_i) = $$ 
$$ = -p log(\frac{q}{p}) + \sum \limits_{i=1}^{n} (- p_i + q_i) \ge (-p (\frac{q}{p} - 1) + \sum \limits_{i=1}^{n}(- p_i + q_i) = 0$$ 

Таким образом, почти за бесплатно получаем утверждение о том, что расстояние Кульбака-Лейблера действительно всегда положительна. 

Условие $D(p, q) = 0 \leftrightarrow p=q$ так же следует из того, что равенства в неравенстве выше выполняются только при равенстве $p_i = q_i, i=\overline{1, n}$. 

Поэтому $D(p, q) = 0 \rightarrow p=q$. Стрелка в обатную сторону тривиально проверяется подстановкой.



Или даже проще можно просто расписать $lnx \le x -1$,  их этой записи сразу следует просто что угодно 

# Задача №5

Пусть $x-$ действительнозначная переменная, принимающая конечный набор значений $a_1 < a_2 < \cdots < a_n$ с вероятностями $\mathbb{P}(x =a_i) = p_i$. Оцените выпуклость и вогнутость следующих функций от $p$ на множестве $S = \left\{p : \sum \limits_{i=1}^{n} p_i = 1, p_i \ge 0\right\}$

* $\mathbb{E}x$
* $\mathbb{P}\{x \ge \alpha\}$
* $\mathbb{P}\{\alpha \le x \le \beta\}$
* $\sum \limits_{i=1}^{n} p_i \log{p_i}$
* $\mathbb{V}x = \mathbb{E}(x - \mathbb{E}x)^2$
* $\mathbf{quantile}(x) = \inf \{\beta \mid \mathbb{P}\{x \le \beta\} \ge 0.25\}$

Решение: 

$\textbf{1)}$ выпуклость и вогнутость первой функции очевидна, потому что она представляет из себя просто некоторую линейную функцию от $p_i, i = \overline{1,n}$

$\textbf{2)}$ Просто подумаем, что из себя вообще эта функция представляет. По сути, мы сначала должны найти такое минимальное $a_i$, что для него будет верно $a_i \ge \alpha$. Таким образом $\mathbb{P}\{x \ge \alpha\}$ запишется как $\sum \limits_{k=i}^{n} p_k$. А эта функция будет линейной функций на выпуклом множестве $\rightarrow$ она будет и выпуклой и вогнутой.

$\textbf{3)}$ Аналогичная ситуация как и в предыдущем пункте. Функция будет записить только от вероятностей некоторых значений, которые лежет  $\alpha \le a_i \le \beta \rightarrow$ она так же будет линейной функций на выпуклом множестве $\rightarrow$ она буде выпуклой и вогнутой

$\textbf{4)}$ Из дифференциального критерия второго порядка получаем, что $f(x) = x \log{x}$ - выпуклая на $S$. Тогда $\sum \limits_{i=1}^{n} p_i \log{p_i}$ будет выпуклой как сумма выпуклых функций с положительными коэффициентами. То, что данная функция не будет вогнутой видно из гессиана функции $-\sum \limits_{i=1}^{n} p_i \log{p_i}$, которая имеет вид $diag\{\frac{-1}{p_i}, i=\overline{1,n}\}$

$\textbf{5)}$ Здесь придется чуть-чуть подумать. Сначала вспомним, что дисперсию можно представить в виде: 

$$\mathbb{V}x = \mathbb{E}x^2 - (\mathbb{E}x)^2 = \sum_i a_i^2 \cdot p_i - (\sum_i a_i \cdot p_i)^2$$

Если честно посчитать гессиан от данной функции, то получим
$$Hess_{\mathbb{V}x } = -2a \cdot a^T$$
Данная матрица будет отрицательно определена $\rightarrow$ по дифференциальному критерию второго порядка получаем, что дисперсия будет вогнутой функцией.

$\textbf{6)}$ Данная функция не будет ни выпуклой, ни вогнутой. Рассмотрим простой пример: 
 зададимся некоторыми распределениями $p_1, p_2$, которые по сути представляют из себя n-мерные вектора. Выберем их так, что в $p_1$ вся вероятность приходится на $a_i$, а в $p_2$ вся вероятность приходится. Очевидно, что для этих распределений $\frac{1}{4}$-квантилями являются сами значения $a_i, a_j$
 
 Отсюда получаем, что 
 
 $$\lambda \cdot \mbox{quantile}(p_1) + (1-\lambda)\cdot \mbox{quantile}(p_2) = \lambda a_i + (1-\lambda) a_j$$
 $$a_i \le \lambda a_i + (1-\lambda) a_j \le a_j$$
 Тогда рассмотрим 
 $$\mbox{quantile}(\lambda p_1 + (1-\lambda) p_2) \mid_{\lambda=0.25} = a_i$$
 $$\mbox{quantile}(\lambda p_1 + (1-\lambda) p_2) \mid_{\lambda=0.01} = a_j$$
 

Таким образом, можно подобрать такие $p_1, p_2, \lambda \in [0, 1]$, что 
$$\lambda \cdot \mbox{quantile}(p_1) + (1-\lambda)\cdot \mbox{quantile}(p_2) \le \mbox{quantile}(\lambda p_1 + (1-\lambda) p_2)$$ 
и 
$$\lambda \cdot \mbox{quantile}(p_1) + (1-\lambda)\cdot \mbox{quantile}(p_2) \ge \mbox{quantile}(\lambda p_1 + (1-\lambda) p_2)$$ 

Отсюда следует, что данная функция не выпуклася и не вогнутая.
