# Задача 1
$f(x) = e^x - 1$, тогда $f''(x) = e^x > 0$ на $R$, значит $f$ выпукла на $R$.

$f(x_1, x_2) = x_1x_2$. 
$$
f''(x) = 
\begin{pmatrix}
0 & 1 \\
1 & 0 \\
\end{pmatrix}.
$$

Эта форма не является ни положительно определенной, ни отрицательно. Поэтому функция не выпукла.

$f(x_1, x_2) = 1/(x_1 x_2)$.
$$
f''(x) = 
\begin{pmatrix}
2/(x_1^3x_2) & 1/(x_1^2x_2^2) \\
1/(x_1^2x_2^2) & 2/(x_1x_2^3) \\
\end{pmatrix}.
$$

$d_1(f''(x)) = 2/(x_1^3x_2) > 0$ и  $d_2(f''(x))= \det(f''(x)) = 3/(x_1^4x_2^4) > 0$ на $R^2_{++}$. Значит $f''(x)$ положительно определена и значит $f$ выпукла на $R^2_{++}$

# Задача 2
Множество $S = \{x \in R^n_{++} \mid \prod\limits_{k = 1}^{n} x_k \ge 1\}$ является надграфиком функции 
$$
f(x_1, \ldots, x_{n-1}) = \frac{1}{\prod_{k = 1}^{n-1} x_k}.
$$

$$
\quad (f'')_{ii} = \frac{2}{x_i^3\prod_{k = 1, k \ne i}^{n-1} x_k}.
$$

$$
(f'')_{ij} = \frac{1}{x_i^2x_j^2\prod_{k = 1, k \ne i, j}^{n-1} x_k}, \quad (i \ne j).
$$

$$
d_k(f'') = C \cdot \det A_k,
$$

где $C$ $-$ некоторая функция обратная к одночлену от $x_1, \ldots x_{n-1}$, а 

$$
A_k = 
\begin{pmatrix}
2 & 1 & 1 & \ldots & 1 \\
1 & 2 & 1 & \ldots & 1 \\
1 & 1 & 2 & \ldots & 1 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & 1 & 1 & \ldots & 2
\end{pmatrix}_{\ \ k \times k}.
$$

Можно заметить, что даже матрица

$$
B_k = 
\begin{pmatrix}
1/2 & 1 & 1 & \ldots & 1 \\
1 & 1/2 & 1 & \ldots & 1 \\
1 & 1 & 1/2 & \ldots & 1 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & 1 & 1 & \ldots & 1/2
\end{pmatrix}_{\ \ k \times k}
$$

положительно определена (квадратичная форма равна полному квадрату), значит и $A_k$ положительна определена, то есть ее определитель больше нуля, значит все главные миноры $f''$ больше нуля, значит $f''$ положительно определена, значит $f$ выпукла, значит множество $S$ выпукло как надграфик $f$.


# Задача 3

Докажем, что $(\det X)^{1/n}$ вогнута. Покажем, что для любой $X \in S^{n}_{++}$ и $V \in R^{n \times n}$ функция $(\det (X + Vt))^{1/n}$ вогнута на множестве $\{t \in R \mid X + Vt \in S^n_{++}\}$.

$$
(\det (X + Vt))^{1/n} = (\det X)^{1/n} (\det (I + tX^{-1/2}VX^{-1/2}))^{1/n} = (\det X)^{1/n} \left(\prod\limits_{i=1}^n (1 + t \lambda_i)\right)^{1/n}.
$$

Матрица $X + Vt$ положительно определена, поэтому матрица $I + tX^{-1/2}VX^{-1/2} = X^{-1/2}(X + Vt){X^{-1/2}}^T$ также положительно определена. Значит все ее собственные числа $1 + t\lambda_i > 0$. Нам нужно доказать неравенство

$$
\left(\prod\limits_{i=1}^n (1 + \theta t_1 \lambda_i + (1-\theta) t_2 \lambda_i)\right)^{\ \ \ \ \ 1/n} \ge \theta \left(\prod\limits_{i=1}^n (1 + t_1\lambda_i)\right)^{1/n} + (1 - \theta) \left(\prod\limits_{i=1}^n (1 + t_2\lambda_i)\right)^{1/n}.
$$

Оно следует из неравенства

$$
\left(\prod\limits_{i=1}^n (\theta x_i + (1-\theta) y_i)\right)^{\ \ \ \ 1/n} \ge \theta \left(\prod\limits_{i=1}^n x_i\right)^{1/n} + (1 - \theta) \left(\prod\limits_{i=1}^n y_i\right)^{1/n},
$$

которое верно для положительных $x_i$ и $y_i$ и $\theta \in [0, 1]$ в силу вогнутости на $R^n_{++}$ функции

$$
F(x) = \left(\prod\limits_{i=1}^n x_i\right)^{1/n},
$$

которую можно доказать по дифференциальному критерию второго порядка.

# Задача 4
$f(p) = \sum p_i \log p_i$, $\dfrac{\partial f}{\partial x_k} = \log p_k + 1$.

$$
f(p) - f(q) - \nabla f(q)^T (p-q) = \sum p_i \log p_i - q_i \log q_i - (\log q_i + 1)(p_i - q_i) = \sum p_i \log(p_i / q_i) - p_i + q_i = D(p, q).
$$

Докажем, что $f(p)$ сильно выпукла в сколь угодно большой $(\forall M > 0)$ области $D_M = B_M(0) \cap R^n_{++}$. 

$$
f''(p) = 
\begin{pmatrix}
1/p_1 & 0 & 0 & \ldots & 0 \\
0 & 1/p_2 & 0 & \ldots & 0 \\
0 & 0 & 1/p_3 & \ldots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & \ldots & 1/p_n
\end{pmatrix} \succeq I \cdot \min\limits_i \inf\limits_{D_M} \frac{1}{p_i} = \mu_M I.
$$

Отсюда следует, что $\forall p, q \in R^n_{++}$, существует некоторое $M$ такое, что

$$
D(p, q) = f(p) - f(q) - \nabla f(q)^T (p-q) \ge \frac{\mu_M}{2} \|p-q\|^2 > 0,
$$

при $p \ne q$. При $p = q$, $D(p, q) = 0$ и мы получаем все утверждения, которые хотели.

# Задача 5

Сначала докажем, что если $f(x)$ линейна на $S$, то есть $\forall x_1, x_2 \in S$ и $\forall \alpha, \beta \in R$ выполнено $f(\alpha x_1 + \beta x_2) = \alpha f(x_1) + \beta f(x_2)$, то она выпукла и вогнута на $S$.

$$
f(\theta x_1 + (1 - \theta) x_2) = \theta f(x_1) + (1 - \theta) f(x_2), \quad \forall x_1, x_2 \in S, \theta \in [0, 1].
$$

Поэтому выполнены оба нестрогих неравенства, что и требовалось.

1. $\mathbb{E} x = \sum p_i a_i$ линейна на вероятностном симплексе, значит выпукла и вогнута.

2. $P(x \ge \alpha) = \sum\limits_{i = k}^{n} p_i$, где $k$ - минимальный номер такой, что $a_k \ge \alpha$. Функция тоже линейна, поэтому выпукла и вогнута.

3. Абсолютно аналогично предыдущему пункту.

4. Функция $f$ из задачи $4$, суженная на вероятностный симплекс выпукла (даже сильно).

5. $\mathbb{V} x = \mathbb{E} x^2 - (\mathbb{E} x)^2 = \sum p_ia_i^2 - (\sum p_i a_i)^2 = f(p)$.
$$
f''(p) = -2
\begin{pmatrix}
a_1^2 & a_1a_2 & \ldots & a_1a_n \\
a_1a_2 & a_2^2 & \ldots & a_2a_n \\
\vdots & \vdots & \ddots & \vdots \\
a_1a_n & a_2a_n & \ldots & a_n^2
\end{pmatrix} \prec 0.
$$
Значит $\mathbb{V} x$ вогнута и не выпукла.

6. Пусть $p_1 = (0 \ldots 0 1)$, $p_2 = (1 0 \ldots 0)$. Тогда, если $f(p) = {\bf quartile}(x)$, то $f(p_1) = a_n$, $f(p_2) = a_1$. Если $1 > \theta \ge 0.25$, то $f(\theta p_1 + (1 - \theta) p_2) = a_1 < \theta a_1 + (1 - \theta) a_n$, а если $0 < \theta < 0.25$, то $f(\theta p_1 + (1 - \theta) p_2) = a_n > \theta a_1 + (1 - \theta) a_n$. Значит $f$ не выпукла и не вогнута на $S$.
